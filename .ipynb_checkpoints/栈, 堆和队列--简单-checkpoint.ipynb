```python
"""
20. 有效的括号
https://leetcode-cn.com/problems/valid-parentheses/
解题思路:
  1. 使用堆栈思想. 如果是左括号"({[", 则将字符压入stack中.
  2. 否则为右括号, 如果stack为空, 或者字符不匹配stack的栈顶元素, 则return False
  3. 最后, 判断stack是否为空.
"""
def isValid(s: str) -> bool:
    # 以右括号作为key, 左括号作为value
    mapping = {')': '(', '}': '{', ']': '['}
    # 堆栈存储压入的左括号
    stack = []
    for _s in s:
        # 如果是左括号, 则压入堆栈中
        if _s in '({[':
            stack.append(_s)
        # 否则为右括号: 1. 如果堆栈为空, 说明右括号没有匹配的括号, 则为非法. 2. 如果堆栈弹出的括号和右括号不匹配, 说明非法.
        elif not stack or stack.pop() != mapping[_s]:
            return False
    return not stack

for s in ["()", "()[]{}", "(]", "([)]", "{[]}"]:
    print(isValid(s))
```

```python
"""
155. 最小栈
https://leetcode-cn.com/problems/min-stack/
解题思路
  1. 定义两个堆栈: stack, 用于存储元素; minStack, 用于存储最小值.
  2. push操作时候, stack正常压入; 但对于minStack来说, 需要判断元素和minStack[-1]的大小, 小的值压入minStack.
  3. pop则删除stack/minStack的栈顶元素; top则获取stack的栈顶元素, getMin则获取minStack的栈顶元素.
"""
class MinStack:
    def __init__(self):
        # 两个堆栈, stack存储元素, minStack存储最小值
        self.stack, self.minStack = [], []
    
    def push(self, x: int) -> None:
        # 将x和minStack[-1]中的最小值压入minStack中
        if not self.minStack:
            self.minStack.append(x)
        else:
            self.minStack.append(min(x, self.minStack[-1]))
        self.stack.append(x)
            
    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()

    def top(self) -> int:
        if self.stack:
            return self.stack[-1]
        raise Exception('stack is empty')

    def getMin(self) -> int:
        if self.minStack:
            return self.minStack[-1]
        raise Exception('min stack is empty')
            
# 简单测试用例
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
# -3
print(minStack.getMin())
minStack.pop()
# 0
print(minStack.top())
# -2
print(minStack.getMin())

"""
LeetCode上的优质答案: https://leetcode.com/problems/min-stack/discuss/342055/Python-48ms-O(1)-for-all-operation-with-algorithm-explanation
解题思路
1. 假设压入a, b, c, 但是a是最小值, 则只要记录其: 长度 + a的组合即可.
"""

class MinStack1:
    def __init__(self):
        self.stack, self.minStack = [], [[-1, 2**32 - 1]]
    
    def push(self, x: int) -> None:
        self.stack.append(x)
        if x < self.minStack[-1][1]:
            self.minStack.append([len(self.stack), x])
    
    def pop(self) -> None:
        self.stack.pop()
        if len(self.stack) < self.minStack[-1][0]:
            self.minStack.pop()
    
    def top(self) -> int:
        return self.stack[-1]
    
    def getMin(self) -> int:
        return self.minStack[-1][1]
    
# 简单测试用例
minStack1 = MinStack1()
minStack1.push(-2)
minStack1.push(0)
minStack1.push(-3)
# -3
print(minStack1.getMin())
minStack1.pop()
# 0
print(minStack1.top())
# -2
print(minStack1.getMin())
```

```python
"""
225. 用队列实现栈
https://leetcode-cn.com/problems/implement-stack-using-queues/

解题思路
  1. 队列是先进先出, 栈是先进后出
  2. 使用两个队列q1, q2. 执行push时候直接插入q1中, pop时候需要将q1的数据复制到q2中, 直到剩下最后一个元素.
"""
class MyStack:
    def __init__(self):
        # 定义两个队列, q1存储元素, q2备份
        self.q1, self.q2 = [], []
    
    def push(self, x: int) -> None:
        self.q1.append(x)
        
    def pop(self) -> int:
        # 除了队列末尾的元素外, 将q1的所有元素全部写入到q2中
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        # 获取队列末尾的元素
        v = self.q1[0]
        # 将备份的q2赋值给q1, 并且置空q2
        self.q1, self.q2 = self.q2, []
        # 返回队列末尾的元素
        return v
    
    def top(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        # 获取队列末尾的元素
        v = self.q1[0]
        # 将末尾元素写入到备份q2中
        self.q2.append(v)
        # 将备份的q2赋值给q1, 并且置空q2
        self.q1, self.q2 = self.q2, []
        # 返回队列末尾的元素
        return v
    
    def empty(self) -> bool:
        return not self.q1
    
stack = MyStack()
stack.push(1)
stack.push(2)
# 2
print(stack.top())
# 2
print(stack.pop())
# False
print(stack.empty())
```

```python
"""
232. 用栈实现队列
https://leetcode-cn.com/problems/implement-queue-using-stacks/

解题思路
  1. 定义两个堆栈, 将值压入s1后, 在不断pop到s2中, 则形成队列.
  2. 可以在pop, peek时候执行s1 --> s2的操作.
"""
class MyQueue:
    def __init__(self):
        # 定义两个堆栈, 一个用于压入, 一个用于形成倒置的堆栈
        self.s1, self.s2 = [], []
        
    def push(self, x: int) -> None:
        # 将元素压入s1中
        self.s1.append(x)
        
    def pop(self) -> int:
        # 如果s2为空, 则将s1的元素弹出, 压入s2中. 这样s2就是s1的倒置
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        # 弹出s2的栈顶元素(即s1的头元素)
        return self.s2.pop()
    
    def peek(self) -> int:
        # 如果s2为空, 则将s1的元素弹出, 压入s2中. 这样s2就是s1的倒置
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        # 获取s2的栈顶元素
        return self.s2[-1]
    
    def empty(self) -> bool:
        return not self.s1 and not self.s2
    
# 测试用例
queue = MyQueue()
queue.push(1)
queue.push(2)
# 1
print(queue.peek())
# 1
print(queue.pop())
# False
print(queue.empty())
```

```python
"""
496. 下一个更大元素I
https://leetcode-cn.com/problems/next-greater-element-i/

解题思路
  1. 循环nums1, 然后一一比较
"""
def nextGreaterElement(nums1, nums2):
    r = []
    for n in nums1:
        index = nums2.index(n)
        for i in range(index + 1, len(nums2)):
            if nums2[i] > n:
                r.append(nums2[i])
                break
        else:
            r.append(-1)
            
    return r

# 简单测试
# [-1, 3, -1]
print(nextGreaterElement([4,1,2],[1,3,4,2]))
# [3, -1]
print(nextGreaterElement([2,4],[1,2,3,4]))

"""
LeetCode上的优质答案:
https://leetcode.com/problems/next-greater-element-i/discuss/337669/easy-peasy-python-solution-using-stack-just-look-for-the-bigger-number-from-the-end-of-nums2

解题思路
  1. 定义map, 用于存储nums1中每个值对应的下一个更大元素
  2. 定义堆栈st, 将nums2从后往前遍历, 不断将大的值压入栈中, 然后将st[-1], 即最小值赋值给map.
"""
def nextGreaterElement1(nums1, nums2):
    # 定义堆栈st存储比当前元素大的值, map存储元素对应的大的值
    st, mp = [], {}
    # 对nums2倒序遍历
    for i in range(len(nums2) - 1, -1, -1):
        num = nums2[i]
        # 如果堆栈为空, 说明没有元素比当前的元素大, 则将当前元素压入到堆栈中, 并且设置mp对应的值为-1
        if not st:
            st.append(num)
            mp[num] = -1
        else:
            # 不断将当前元素值和堆栈的最后一个元素比较(最后一个元素即堆栈中最小的元素)
            while st and num > st[-1]:
                st.pop()
            # 如果堆栈为空, 则说明没有值比当前值大; 否则比当前值大的, 就是堆栈栈顶元素(注意nums2是倒序遍历)
            mp[num] = -1 if not st else st[-1]
            # 将当前值压入到堆栈中
            st.append(num)
    r = []
    # 遍历nums1, 从mp中读取大于当前值的值
    for n in nums1:
        r.append(mp.get(n, -1))
    return r
# 简单测试
# [-1, 3, -1]
print(nextGreaterElement1([4,1,2],[1,3,4,2]))
# [3, -1]
print(nextGreaterElement1([2,4],[1,2,3,4]))
```

```python
"""
682. 棒球比赛
https://leetcode-cn.com/problems/baseball-game/submissions/

解题思路
  1. 使用堆栈即可.
"""
def calPoints(ops) -> int:
    r = []
    for _ops in ops:
        if _ops == '+':
            r.append(r[-1] + r[-2])
        elif _ops == 'D':
            r.append(r[-1] * 2)
        elif _ops == 'C':
            r.pop()
        else:
            r.append(int(_ops))
            
    return sum(r)

# 简单测试
print(calPoints(["5","2","C","D","+"]))
```

```python
"""
703. 数据流中的第K大元素
"""
import heapq

class KthLargest:

    def __init__(self, k: int, nums):
        # 这里必须使用heapq, 否则时间永远超时
        self.pool = heapq.nlargest(k, nums)
        heapq.heapify(self.pool)
        self.k = k

    def add(self, val: int) -> int:
        if len(self.pool) < self.k:
            heapq.heappush(self.pool,val)
        else:
            heapq.heappushpop(self.pool,val)
        return self.pool[0]
```

```python
"""
844. 比较含退格的字符串
https://leetcode-cn.com/problems/backspace-string-compare/
解题思路
  1. 使用堆栈即可.
"""
def backspaceCompare(S: str, T: str) -> bool:
    s1, s2 = [], []
    for s in S:
        if s == '#' and s1:
            s1.pop()
        elif s != '#':
            s1.append(s)
    for t in T:
        if t == '#' and s2:
            s2.pop()
        elif t != '#':
            s2.append(t)
    return s1 == s2

# 简单测试
# True
print(backspaceCompare("ab#c", "ad#c"))
```

```python
"""
933. 最近的请求次数
https://leetcode-cn.com/problems/number-of-recent-calls/
"""
class RecentCounter:
    
    def __init__(self):
        self.times = []
        
    def ping(self, t: int) -> int:
        self.times.append(t)
        i = 0
        while i < len(self.times) - 1:
            if self.times[i] >= self.times[-1] - 3000:
                break
            i += 1
        self.times = self.times[i:]
        return len(self.times)
```

```python
"""
1021. 删除最外层的括号
https://leetcode-cn.com/problems/remove-outermost-parentheses/

解题思路
  1. 使用堆栈存储括号. 并使用left, right分别代表左括号, 右括号的数量.
  2. 如果left == right, 则说明分解成功, 则去除括号即可.
"""
def removeOuterParentheses(S: str) -> str:
    r, stack, left, right = [], [], 0, 0
    for s in S:
        stack.append(s)
        if s == '(':
            left += 1
        else:
            right += 1
        # 说明存在括号包围, 则去除括号
        if left == right:
            r.append(''.join(stack[1:-1]))
            stack, left, right = [], 0, 0
    
    return ''.join(r)

# 简单测试
# "()()()"
print(removeOuterParentheses("(()())(())"))
# "()()()()(())"
print(removeOuterParentheses("(()())(())(()(()))"))
# ""
print(removeOuterParentheses("()()"))
```

```python
"""
1046. 最后一块石头的重量
"""
def lastStoneWeight(stones):
    stones.sort()
    if len(stones) == 1:
        return stones[0]
    if len(stones) == 2:
        return stones[1] - stones[0]
    return lastStoneWeight(stones[:-2] + [stones[-1] - stones[-2]])

print(lastStoneWeight([4,3,7,6,8,9,2]))
```

```python
"""
1047. 删除字符串中的所有相邻重复项
https://leetcode-cn.com/problems/remove-all-adjacent-duplicates-in-string/
解题思路
  1. 使用堆栈, 每次插入数值, 均跟栈顶元素比较. 如果相等, 则弹出栈顶元素即可.
"""
def removeDuplicates(S: str) -> str:
    stack = []
    for s in S:
        if stack and stack[-1] == s:
            stack.pop()
        else:
            stack.append(s)
    
    return ''.join(stack)

# 简单测试
# ca
print(removeDuplicates("abbaca"))       
```
