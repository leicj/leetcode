```python
class Node:
    def __init__(self, val, nex=None, rand=None):
        self.val = val
        self.next = nex
        self.rand = rand
        
class DoubleNode:
    def __init__(self, val):
        self.val = val
        self.last = self.next = None
        
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None
        
def array2node(arr, isCircle=False):
    if not arr:
        return None
    node = Node(arr[0])
    head = node
    for i in range(1, len(arr)):
        node.next = Node(arr[i])
        node = node.next
    # 尾部指针指向头部指针
    if isCircle:
        node.next = head
    return head

def node2array(head):
    if not head:
        return []
    r, root = [], head
    while root:
        r.append(root.val)
        root = root.next
    return r

def array2tree(arr):
    n = arr.pop(0)
    root = None if n is None else TreeNode(n)
    queue = [root]
    while arr and queue:
        node = queue.pop(0)
        if node is None:
            continue
        
        n1 = arr.pop(0)
        node.left = None if n1 is None else TreeNode(n1)
        queue.append(node.left)
        if arr:
            n2 = arr.pop(0)
            node.right = None if n2 is None else TreeNode(n2)
            queue.append(node.right)
    return root
```

```python
"""
2. 两数相加
"""
def addTwoNumbers(l1, l2):
    r, ca = Node(None), 0
    root = r
    while l1 and l2:
        v = l1.val + l2.val + ca
        r.next = Node(v % 10)
        r = r.next
        ca = v // 10
        l1, l2 = l1.next, l2.next
    while l1:
        v = l1.val + ca
        r.next = Node(v % 10)
        r = r.next
        ca = v // 10
        l1 = l1.next
    while l2:
        v = l2.val + ca
        r.next = Node(v % 10)
        r = r.next
        ca = v // 10
        l2 = l2.next
    if ca == 1:
        r.next = Node(1)
    return root.next

l1, l2 = array2node([2,4,3]), array2node([5,6,4])
print(node2array(addTwoNumbers(l1, l2)))
```

```python
"""
19. 删除链表的倒数第N个节点
"""
def removeNthFromEnd(head, n):
    fast, slow, pre = head, head, head
    while n:
        fast = fast.next
        n -= 1
    if not fast:
        return head.next
    while fast and slow:
        pre = slow
        fast = fast.next
        slow = slow.next
    pre.next = pre.next.next
    return head

head1, head2 = array2node([1,2,3,4,5]), array2node([1,2,3,4,5])
print(node2array(removeNthFromEnd(head1, 2)))
print(node2array(removeNthFromEnd(head2, 5)))
```

```python
"""
24. 两两交换链表中的节点
"""
def swapPairs(head):
    if not head or not head.next:
        return head
    root = head.next
    pre, cur, nex = None, None, None
    while head and head.next:
        nex = head.next.next
        cur = head.next
        if pre:
            pre.next = cur
        cur.next = head
        head.next = nex
        pre = head
        head = nex
    return root

print(node2array(swapPairs(array2node([1,2,3,4]))))
print(node2array(swapPairs(array2node([1,2,3,4,5]))))
        
```

```python
"""
61. 旋转链表
题目: 给定一个链表, 旋转链表, 将链表每个节点向右移动k个位置.
1->2->3->4->5->None, k = 2;
输出: 4->5->1->2->3->None
"""
def rotateRight(head, k):
    if not head:
        return head
    root = Node(None)
    root.next = head
    clone, length, slow, fast = head, 0, head, head
    # 获取链表的长度
    while clone:
        clone = clone.next
        length += 1
    k = k % length
    
    # 定位到第k个节点
    while fast and k:
        fast = fast.next
        k -= 1
    # 将slow定位到倒数k个节点的前一个节点, fast定位到最后一个节点(因为slow和fast相距k个节点)
    while fast.next:
        slow, fast = slow.next, fast.next
    fast.next = root.next
    root.next = slow.next
    slow.next = None
    return root.next

head = array2node([1,2,3,4,5])
print(node2array(rotateRight(head, 2)))
```

```python
"""
82. 删除排序链表中的重复元素II
1->2->3->3->4->4->5->None
输出: 1->2->5->None
"""
def deleteDuplicates(head):
    if not head:
        return head
    root = Node(None)
    root.next = head
    flag, clone = False, root
    
    while head.next:
        if head.val == head.next.val:
            head.next = head.next.next
            flag = True
        else:
            if flag:
                clone.next = head.next
                flag = False
            else:
                clone = head
            head = head.next
    if flag:
        clone.next = None
    return root.next

head = array2node([1,2,3,3,4,4,5])
print(node2array(deleteDuplicates(head)))
```

```python
"""
86. 分隔链表
解题思路:
1. 使用额外空间存储大小节点.
2. 使用两个大小指针遍历
"""
def partition(head, x):
    if not head:
        return head
    small, large = Node(None), Node(None)
    smallHead, largeHead = small, large
    root = head
    while root:
        if root.val < x:
            small.next = Node(root.val)
            small = small.next
        else:
            large.next = Node(root.val)
            large = large.next
        root = root.next
    small.next = largeHead.next
    return smallHead.next

def partition2(head, x):
    sh, st, bh, bt = None, None, None, None
    while head:
        nex = head.next
        head.next = None
        if head.val < x:
            if not sh:
                sh, st = head, head
            else:
                st.next = head
                st = st.next
        else:
            if not bh:
                bh, bt = head, head
            else:
                bt.next = head
                bt = bt.next
        head = nex
    
    if st:
        st.next = bh
    return sh or bh

head = array2node([1,4,3,2,5,2])
print(node2array(partition(head, 3)))
print(node2array(partition2(head, 3)))
```

```python
"""
92. 反转链表II
题目: 反转从位置m到n的链表.
"""
def reverseBetween(head, m, n):
    root = Node(None)
    root.next = head
    slow, fast, slowP = root, root, None
    while m > 0 and n > 0:
        if m == 1:
            slowP = slow
        slow, fast = slow.next, fast.next
        m -= 1
        n -= 1
    while n > 0:
        fast = fast.next
        n -= 1
    # 到这步为止, 按照1->2->3->4->5->None, m=2, n=4.
    # 则slowP指向1, slow指向2, fast指向4
    pre = fast.next
    while slow != fast:
        nex = slow.next
        slow.next = pre
        pre = slow
        slow = nex
    slow.next = pre
    slowP.next = slow
    
    return root.next

head = array2node([1,2,3,4,5])
print(node2array(reverseBetween(head, 2, 4)))
```

```python
"""
109. 有序链表转换二叉搜索树
"""
def sortedListToBST(head):
    r = []
    while head:
        r.append(head.val)
        head = head.next
    
    def help(arr):
        length = len(arr)
        if length == 0:
            return None
        mid = length // 2
        root = TreeNode(arr[mid])
        root.left, root.right = help(arr[0:mid]), help(arr[mid + 1:])
        return root
    return help(arr)
```

```python
"""
138. 复制带随机指针的链表
"""
def copyRandomList(head):
    d, cur = {}, head
    while cur:
        d[cur] = Node(cur.val)
        cur = cur.next
    cur = head
    while cur:
        d[cur].next = d.get(cur.next, None)
        d[cur].rand = d.get(cur.rand, None)
        cur = cur.next
    return d.get(head, None)
```

```python
"""
142. 环形链表II
"""
def detectCycle(head):
    if not head or not head.next or not head.next.next:
        return None
    slow, fast = head.next, head.next.next
    while slow and fast and fast.next and slow != fast:
        slow = slow.next
        fast = fast.next.next
    fast = head
    while slow and fast and slow != fast:
        slow, fast = slow.next, fast.next
    return slow
```

```python
"""
143. 重排链表
题目:
给定一个单链表 L：L0→L1→…→Ln-1→Ln ，
将其重新排列后变为： L0→Ln→L1→Ln-1→L2→Ln-2→…
"""
def reorderList(head):
    if not head:
        return head
    root, tail = head, head
    while tail and tail.next:
        tail.next.parent = tail
        tail = tail.next
    while root != tail and root.next != tail:
        nex = root.next
        tail.next = nex
        root.next = tail
        root = nex
        tail = tail.parent
        tail.next = None
```

```python
"""
147. 对链表进行插入排序
"""
def insertionSortList(head):
    root = Node(float("-inf"), head)
    cur = root
    
    while cur.next:
        # 说明前cur.next个节点已经有序了, 则无需排序
        if cur.next.val >= cur.val:
            cur = cur.next
            continue
        # 将要被插入到前面有序链表的节点
        node = cur.next
        # 将此节点删除, 等待插入
        cur.next = cur.next.next
        start, prev = root, root
        while start != cur.next:
            if node.val <= start.val:
                prev.next = node
                node.next = start
                break
            prev = start
            start = start.next
    
    return root.next

head = array2node([6,5,3,1,8,7,2,4])
print(node2array(insertionSortList(head)))
```

```python
"""
148. 排序链表
题目: 在O(n log n)时间复杂度和常数级空间复杂度下, 对链表进行排序.
"""
def sortList(head):
    if not head or not head.next:
        return head
    # 将链表分隔成两个部分
    slow, fast = head, head.next
    while fast and fast.next:
        fast = fast.next.next
        slow = slow.next
    second = slow.next
    slow.next = None
    left, right = sortList(head), sortList(second)
    return merge(left, right)

def merge(left, right):
    if not left or not right:
        return left or right
    if left.val > right.val:
        left, right = right, left
    # 这里, left和right均是已经排序的链表
    head = prev = left
    left = left.next
    while left and right:
        if left.val < right.val:
            left = left.next
        else:
            nex = prev.next
            prev.next = right
            tmp = right.next
            right.next = nex
            right = tmp
        prev = prev.next
    prev.next = left or right
    return head

head = array2node([2,4,5,2,3,8,7,5,6,3])
print(node2array(sortList(head)))
```

```python
"""
328. 奇偶链表
"""
def oddEvenList(head):
    if not head:
        return head
    n1, n2, n3 = head, head.next, head.next
    while n1 and n1.next and n2 and n2.next:
        n1.next = n1.next.next
        n1 = n1.next
        n2.next = n2.next.next
        n2 = n2.next
    n1.next = n3
    return head

head = array2node([1,2,3,4,5,6,7,8])
print(node2array(oddEvenList(head)))
```

```python
"""
430. 扁平化多级双向链表
"""
def flatten(head):
    if not head:
        return head
    queue = []
    def dfs(node):
        if not node:
            return None
        queue.append(node)
        dfs(node.child)
        dfs(node.next)
    dfs(head)
    
    for i in range(1, len(queue)):
        first, second = queue[i - 1], queue[i]
        first.next = second
        second.prev = first
        first.child, second.child = None, None
    
    return queue[0]
```

```python
"""
445. 两数相加
7->2->4->3 + 5->6->4 = 7->8->0->7
"""
def addTwoNumbers(l1, l2):
    # 使用堆栈
    s1, s2 = [], []
    while l1:
        s1.append(l1.val)
        l1 = l1.next
    while l2:
        s2.append(l2.val)
        l2 = l2.next
    root = Node(None)
    ac = 0
    while s1 or s2:
        v = ac
        if s1:
            v += s1.pop()
        if s2:
            v += s2.pop()
        root.val = v % 10
        tmp = Node(None)
        tmp.next = root
        root = tmp
        ac = v // 10
    if ac == 1:
        root.val = 1
        return root
    return root.next

l1, l2 = array2node([7,2,4,3]), array2node([5,6,4])
print(node2array(addTwoNumbers(l1, l2)))
l3, l4 = array2node([9,9,9,9]), array2node([1])
print(node2array(addTwoNumbers(l3, l4)))
```

```python
"""
725. 分隔链表
"""
def splitListToParts(root, k):
    r = []
    if not root:
        return [None for _ in range(k)]
    head, first, length = root, root, 0
    while head:
        head = head.next
        length += 1
    # v是每个分隔的平均长度, m是多出来的部分
    v = length // k
    m = length % k
    for i in range(k):
        if v > 0 or m > 0:
            r.append(first)
        else:
            r.append(None)
        prev = None
        for j in range(v):
            prev = first
            first = first.next
        if m > 0:
            prev = first
            first = first.next
            m -= 1
        if prev:
            prev.next = None
    return r
```

```python
"""
817. 链表组件
"""
def numComponents(head, G):
    if not head or not G:
        return 0
    count, flag = 0, False
    while head:
        val = head.val
        if val in G:
            G.remove(val)
            flag = True
        elif flag:
            count += 1
            flag = False
        head = head.next
    
    return count + 1 if flag else count
```

```python
"""
1019. 链表中的下一个更大节点
"""
def nextLargerNodes(head):
    nums = []
    while head:
        nums.append(head.val)
        head = head.next
    res, stack = [0 for _ in range(len(nums))], []
    for i in range(len(nums) * 2 - 1, -1, -1):
        while stack and nums[stack[-1]] <= nums[i % len(nums)]:
            stack.pop()
        res[i % len(nums)] = nums[stack[-1]] if stack else 0
        stack.append(i % len(nums))
    
    res[-1] = 0
    return res

def nextLargerNodes2(head):
    stack, ans = [(float("inf"), -1)], []
    count = -1
    while head:
        ans.append(0)
        count += 1
        
        while head.val > stack[-1][0]:
            ans[stack[-1][1]] = head.val
            stack.pop()
        stack.append((head.val, count))
        head = head.next
    return ans

head1, head2 = array2node([2,7,4,3,5]), array2node([2,1,5])
print(nextLargerNodes(head1))
print(nextLargerNodes(head2))
print(nextLargerNodes2(head1))
print(nextLargerNodes2(head2))
```
